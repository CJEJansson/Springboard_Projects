Application of a regression model may not yield great results, because several of the variables have non-linear relationships. However, it doesn't hurt to verify this. It's also entirely possible that a combination model using regression may prove useful. 

Logistic regression has been chosen because the output is binary, and is essentially a classification problem. Were the variables continuous we could leverage Linear Regression.

```python
#List of Logistic Regression Solvers 
solvers = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']

#Initialize empty list of scores
solver_scores=[]

#Function to calculate ML scores for Logistic Regression 
for param in solvers:
    clr = LogisticRegression(solver = param, max_iter = 10000, multi_class = 'ovr')
    fit = clr.fit(X_train,y_train)
    score = accuracy_score(clr.predict(X_test), y_test)
    solver_scores.append(score)

max_score = max(solver_scores)
max_idx = solver_scores.index(max(solver_scores))
highest_accuracy = solvers[max_idx]

print(' Logistic Regression Accuracy Score Values: ' , solver_scores)
print("Best score: ", max_score)
print("Solver that yields best score =", highest_accuracy)
```
